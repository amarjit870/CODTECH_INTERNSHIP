# USE PYTHON TO FETCH DATA FROM A PUBLIC API (E.G., OPENWEATHERMAP) AND CREATE VISUALIZATIONS USING MATPLOTLIB OR SEABORN

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Fetch data from DummyJSON
url = "https://dummyjson.com/products"
response = requests.get(url)
response.raise_for_status()  # ensure no HTTP errors
data = response.json()
```

```python
data_df = pd.DataFrame(data)
```

```python
data_df
```

```python
data_df['products']
```

```python
# 2. Load into DataFrame
df = pd.DataFrame(data['products'])

# 3. Clean / process data: take top 10 products by rating
top10 = df.nlargest(10, 'rating')
```

```python
df
```

```python
print(top10)
```

```python
# 4. Plot with seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='title', y='rating', data=top10, palette='viridis')

# Rotate x-axis labels vertically
plt.xticks(rotation=90)

plt.title("Top 10 Rated Dummy Products")
plt.xlabel("Rating (out of 5)")
plt.ylabel("Product Title")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 4. Plot with seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='rating', y='title',        # <-- note I swapped x/y to match labels
            data=top10, 
            palette='viridis')

# Rotate y‑axis labels if titles are long
plt.yticks(rotation=0)  

plt.title("Top 10 Rated Dummy Products")
plt.xlabel("Rating (out of 5)")
plt.ylabel("Product Title")
plt.tight_layout()
plt.show()
```

```python
# 2. Extract top 10 rated products
top10 = df.sort_values(by='rating', ascending=False).head(10)
```

```python
top3_price = df.nlargest(3, 'price')
top3_price
```

```python
# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top3_price['price'],
    labels=top3_price['title'],
    autopct='%1.1f%%',  # shows percentage
    startangle=140
)

# Title and hover info (optional: can show price in legend or title)
plt.title("Top 3 Most Expensive Products - Price Distribution")

# Add legend with price info
plt.legend(
    [f"{title} - ${price}" for title, price in zip(top3_price['title'], top3_price['price'])],
    title="Product & Price",
    loc="best"
)

plt.tight_layout()
plt.show()
```
